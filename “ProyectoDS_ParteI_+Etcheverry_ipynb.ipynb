El coeficiente de correlación(p value) indica que hay una correlación negativa entre las dos variables, cuando aumenta la otra decrese.

# A través del siguiente gráfico quiero comprobar si existe una correlación entre la variable 'rank' y la variable 'Animation_score'

A través del siguiente gráfico quiero comprobar si existe una correlación entre la variable 'members' y la variable 'popularity'.

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.scatterplot(ordered_byrank50, x = "popularity", y = "members", hue='genre');
# Reverse the x-axis
plt.gca().invert_xaxis()
plt.xlabel('Popularity')
plt.ylabel('Members')
plt.title('Relación entre Popularity y members top 50')
```

Con estos dos gráficos podemos comprobar que indican una correlación entre la variable **'rank'** con la variable **'Animation_score'**, y la variable **'members'** con la variable **'popularity'**.

Ahora voy a realizar calculos de p_value para tener un valor numerico de la correlación.

```python
# Chequeo por coeficiente de correlación 'p value' entre estas dos variables
df_anime_rank_50 = ordered_byrank50['rank']
df_anime_animation_50 = ordered_byrank50['Animation_score']

# Chequeo for NaN values in el Array
nan_mask = np.isnan(df_anime_rank_50) | np.isnan(df_anime_animation_50)

#   Remuevo los Nan Values de los dos arrays
df_anime_rank_50 = df_anime_rank_50[~nan_mask]
df_anime_animation_50 = df_anime_animation_50[~nan_mask]

#correlation_coefficient = np.corrcoef(df_anime_rank_50 , df_anime_animation_50)[0,1]
#print("El coeficiente de correlación es de : ", correlation_coefficient )
correlation_coefficient, p_value = spearmanr(df_anime_rank_50, df_anime_animation_50)
correlation_coefficient_pearsonsr, p_value = pearsonr(df_anime_rank_50, df_anime_animation_50)
print("Estos son los coeficiente de correlación entre 'Rank' y ' Animation_score'.")
print("El coeficiente de correlación es de:", correlation_coefficient)
print("El coeficiente de correlación es de(pearsonsr):", correlation_coefficient_pearsonsr)
print("El p-value es de:", p_value)
```

```python
df_members = df_anime['members']
df_popularity = df_anime['popularity']

nan_mask = np.isnan(df_members) | np.isnan(df_popularity)

# Remuevo los NaN values en los dos arrays
df_members = df_members[~nan_mask]
df_popularity = df_popularity[~nan_mask]

correlation_coefficient, p_value = spearmanr(df_members, df_popularity)
correlation_coefficient_pearsonsr, p_value = pearsonr(df_members, df_popularity)
print("Estos son los coeficiente de correlación entre 'Members' y ' Popularity'.")
print("El coeficiente de correlación es de:", correlation_coefficient)
print("El coeficiente de correlación es de(pearsonsr):", correlation_coefficient_pearsonsr)
print("El p-value es de:", p_value)

spearmanr(df_anime['popularity'], df_anime['members'])
```

En este caso el coeficiente de relación es -0.96 muy cercano a -1. Como la variable Rank está ordenada de 50 a 1. Siendo 1 el valor más alto.

En su contexto, si la variable "rango" representa una clasificación en la que los valores más bajos indican elementos mejores o mejor clasificados (por ejemplo, una clasificación de 50 a 1 en la que 1 es el mejor), entonces un coeficiente de correlación negativo sugiere una fuerte relación negativa entre "Rank" y "Animation_score".

En este caso:

Un "Rank" más bajo (más cercano a 1) se asocia a una "Animation_score" más alta.
Un "Rank" más alto (cercano a 50) se asocia a una "Animation_score" más baja.
Por lo tanto, a medida que disminuye el "Rank" (pasando de 50 a 1), la "Animation_score" tiende a aumentar, lo que indica que los elementos mejor clasificados tienden a tener puntuaciones de animación más altas. Esto es coherente con la interpretación de una correlación negativa en el contexto de sus datos.

La misma observación podemos tener entre 'members' y 'popularity' de acuerdo al calculo de P-value de Spearmanr el valor es de -0.99. Lo que indica una Correlación practicamente absoluta.

```python
# El primer algoritmo que utilizo va a ser PCA para FEATURE SELECTION.
# mi variable target es 'popularity'
X_numeric= df_anime[['reviewers','Animation_score','favorites','episodes','rank']] #elimino members
#X_numeric_filled = X_numeric.drop(columns=['animeID'])

# Sustituyo los datos NaN con el promomedio o media (mean)
X_numeric_filled = X_numeric.fillna(X_numeric.mean())

# Standarize data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric_filled)

# Separando data entre entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X_scaled,  df_anime['popularity'], test_size=0.2, random_state=42)

# Aplico PCA

pca = PCA(n_components=2 ) # Numeros de componenentes
X_pca = pca.fit_transform(X_train)

# Entrenando modelo (Linear Regression)
model = LinearRegression()
model.fit(X_pca, y_train)

# Tranformando el set utilizando el mismo PCA
X_test_pca = pca.transform(X_test)

# Predicciones sobre el Test
y_pred = model.predict(X_test_pca)

mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (After PCA): {mse:.2f}")
```

```python

# Mi variable target es 'popularity'
X_numeric= df_anime[['reviewers','Animation_score','favorites','episodes','rank']]

y = df_anime['popularity']

# Sustituyo los datos NaN con el promomedio o media (mean)
X_numeric_filled = X_numeric.fillna(X_numeric.mean())

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_numeric_filled, y, test_size=0.2, random_state=42)

# Standarize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric_filled)

# Forward Feature Selection
selected_features = []
remaining_features = list(X_numeric_filled.columns)

while remaining_features:
    best_feature = None
    best_mse = float('inf')

    for feature in remaining_features:
        current_features = selected_features + [feature]
        X_subset = X_train[current_features]

        # Train your model (using Linear Regression as an example)
        model = LinearRegression()
        model.fit(X_subset, y_train)

        # Make predictions on the test set
        y_pred = model.predict(X_test[current_features])

        # Evaluate performance using Mean Squared Error
        mse = mean_squared_error(y_test, y_pred)

        # Update best feature if MSE improves
        if mse < best_mse:
            best_mse = mse
            best_feature = feature

    # Add the best feature to the selected features
    selected_features.append(best_feature)

    # Use a try-except block to handle the case when a feature is not found
    try:
        remaining_features.remove(best_feature)
    except ValueError:
        print(f"Warning: {best_feature} not found in remaining_features.")

    print(f"Selected Features: {selected_features}, MSE: {best_mse}")
```

```python
# Outside the loop, calculate metrics using the final model
# Make predictions on the test set with the model trained using the selected features
y_pred_final = model.predict(X_test[selected_features])

# Calcular metricas

r2_final = r2_score(y_test, y_pred_final)
mae_final = mean_absolute_error(y_test, y_pred_final)
mse_final = mean_squared_error(y_test,y_pred_final)

# Print las metricas adicionales
print(f"R-squared: {r2_final:.2f}")
print(f"Mean Absolute Error: {mae_final:.2f}")
print(f"Mean Squared Error: {mse_final:.2f}")
```
