Ahora un Scatter Plot. Que indica la relación entre la popularidada de un anime y la puntuación de la animación. Con el Dataframe entero.

```python
df_anime_popularity_total = df_anime['popularity']
df_anime_animation_score_total = df_anime['Animation_score']
```

```python
fig, ax = plt.subplots()
ax.scatter(df_anime_animation_score_total ,df_anime_popularity_total, alpha=0.7)
ax.set_title('Relación entre Animation Score y Popularidad')
ax.set_xlabel('Animation Score')
ax.set_ylabel('Popularity')
```

```python
# Ahora voy a calcular el coeficiente de correlación p value entre estas dos variables

# Chequeo for NaN values in el Array
nan_mask = np.isnan(df_anime_popularity_total) | np.isnan(df_anime_animation_score_total)

#   Remuevo los Nan Values de los dos arrays
df_anime_popularity_total = df_anime_popularity_total[~nan_mask]
df_anime_animation_score_total = df_anime_animation_score_total[~nan_mask]

correlation_coefficient = np.corrcoef(df_anime_popularity_total, df_anime_animation_score_total)[0,1]
print("El coeficiente de correlación es de : ", correlation_coefficient )
```

**Interpretación:**
-  Este valor sugiere una correlación negativa moderadamente fuerte entre la popularidad del anime y su Animation Score. Esto implica que, en general, cuando el Animation Score aumenta, la popularidad tiende a disminuir, y viceversa.
- Según este Scatter Plot existe una relación entre la popularidad del anime y su -  Animation Score. Entre un puntaje de 6 y 7.5 o 8, se encuentran la mayoría de los animes con más puntuación y más popularidad.

- Claramente se ven outliers. Hay por lo menos 2 outliers con puntuación menor de 2 y están en un nivel muy alto de popularidad. También son muy pocos los outliers que tienen un puntaje de animación de 10 y con popularidad máxima.

- El coeficiente de correlación indica que hay una correlación negativa entre las dos variables, cuando aumenta la otra decrese.

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.scatterplot(ordered_byrank50, x = "popularity", y = "members", hue='genre');
# Reverse the x-axis
plt.gca().invert_xaxis()
plt.xlabel('Popularity')
plt.ylabel('Members')
plt.title('Relación entre Popularity y members top 50')
```

Con estos dos gráficos podemos comprobar que indican una correlación entre la variable **'rank'** con la variable **'Animation_score'**, y la variable **'members'** con la variable **'popularity'**.

Ahora voy a realizar calculos de p_value para tener un valor numerico de la correlación.

```python
# Chequeo por coeficiente de correlación 'p value' entre estas dos variables
df_anime_rank_50 = ordered_byrank50['rank']
df_anime_animation_50 = ordered_byrank50['Animation_score']

# Chequeo for NaN values in el Array
nan_mask = np.isnan(df_anime_rank_50) | np.isnan(df_anime_animation_50)

#   Remuevo los Nan Values de los dos arrays
df_anime_rank_50 = df_anime_rank_50[~nan_mask]
df_anime_animation_50 = df_anime_animation_50[~nan_mask]

#correlation_coefficient = np.corrcoef(df_anime_rank_50 , df_anime_animation_50)[0,1]
#print("El coeficiente de correlación es de : ", correlation_coefficient )
correlation_coefficient, p_value = spearmanr(df_anime_rank_50, df_anime_animation_50)
correlation_coefficient_pearsonsr, p_value = pearsonr(df_anime_rank_50, df_anime_animation_50)
print("Estos son los coeficiente de correlación entre 'Rank' y ' Animation_score'.")
print("El coeficiente de correlación es de:", correlation_coefficient)
print("El coeficiente de correlación es de(pearsonsr):", correlation_coefficient_pearsonsr)
print("El p-value es de:", p_value)
```

```python
df_members = df_anime['members']
df_popularity = df_anime['popularity']

nan_mask = np.isnan(df_members) | np.isnan(df_popularity)

# Remuevo los NaN values en los dos arrays
df_members = df_members[~nan_mask]
df_popularity = df_popularity[~nan_mask]

correlation_coefficient, p_value = spearmanr(df_members, df_popularity)
correlation_coefficient_pearsonsr, p_value = pearsonr(df_members, df_popularity)
print("Estos son los coeficiente de correlación entre 'Members' y ' Popularity'.")
print("El coeficiente de correlación es de:", correlation_coefficient)
print("El coeficiente de correlación es de(pearsonsr):", correlation_coefficient_pearsonsr)
print("El p-value es de:", p_value)

spearmanr(df_anime['popularity'], df_anime['members'])
```

**Interpretación correlación entre 'Rank' y 'Animation_score':**

- Coeficiente de Correlación de Pearson: Valor: -0.9657726196568992
- Esto indica una correlación negativa extremadamente fuerte. A medida que aumenta el Animation_score, el Rank (suponiendo que un valor más bajo de Rank es mejor) tiende a mejorar significativamente.

- En su contexto, si la variable "rango" representa una clasificación en la que los valores más bajos indican elementos mejores o mejor clasificados (por ejemplo, una clasificación de 50 a 1 en la que 1 es el mejor), entonces un coeficiente de correlación negativo sugiere una fuerte relación negativa entre "Rank" y "Animation_score".

- Un "Rank" más bajo (más cercano a 1) se asocia a una "Animation_score" más alta.Un "Rank" más alto (cercano a 50) se asocia a una "Animation_score" más baja. Por lo tanto, a medida que disminuye el "Rank" (pasando de 50 a 1), la "Animation_score" tiende a aumentar, lo que indica que los elementos mejor clasificados tienden a tener puntuaciones de animación más altas. Esto es coherente con la interpretación de una correlación negativa en el contexto de sus datos.

#Hypertunning utiliando la metodología RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from scipy.stats import randint, uniform

from sklearn.metrics import make_scorer


# Define a function to calculate R-squared score
def r2_scorer(y_true, y_pred):
    return r2_score(y_true, y_pred)

# Make a scorer from the R-squared function
r2_scorer = make_scorer(r2_scorer)

# Definir pipeline

pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('regressor', RandomForestRegressor(random_state=42))

])

# Definir hyperparametros a buscar
param_distributions = {
    'regressor__n_estimators': randint(100,300),
    'regressor__max_depth': [None, 10, 20, 30],
    'regressor__min_samples_split': randint(2, 20),
    'regressor__min_samples_leaf': randint(1, 20),
    'regressor__max_features': ['auto', 'sqrt', 'log2'],
    'regressor__bootstrap': [True]
}

# Crear RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_distributions,
    n_iter=50,
    scoring={'neg_mean_squared_error': 'neg_mean_squared_error', 'r2': r2_scorer},
    refit='r2',  # Specify which metric to use for refitting
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1

)

# Fit el modelo
random_search.fit(X_train_scaled, y_train_imputed)

# Get the best parameters and best score
print('Mejores parametros', random_search.best_params_)
print("Best Negative Mean Squared Error:", random_search.best_score_)
print("Best R-squared Score:", random_search.cv_results_['mean_test_r2'][random_search.best_index_])

# Predicciones
y_pred = random_search.best_estimator_.predict(X_test_scaled)

#Evaluar modelo
mse= mean_squared_error(y_test_imputed,y_pred)
print("Mean Squared Error",mse)

# Evaluate the best model
r2 = r2_score(y_test_imputed, y_pred)
print("R2_score", r2)
```
